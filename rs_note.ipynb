# 推荐系统
***
## 核心算法
***
- 协同过滤算法
- 隐语义模型

## 协同过滤算法
1. 收集用户偏好
  - 用户评分矩阵
2. 找到相似的物品或者用户
  - 计算相似度
    - 欧氏距离
    - 皮尔森距离
    - 余弦距离
  - 选择邻节点
    - 基于数目
    - 基于邻域
3. 计算推荐模型

## 基于用户的协同过滤--节点为用户
1. 对用户提取特征向量
2. 根据特征向量计算相似度
3. 根据相似度推荐

#### 要解决的问题

> 已知用户评分矩阵（通常是稀疏矩阵），推断该矩阵中零元素的值

####  存在的问题
  > 新用户很难找到邻节点
  > 对于一个物品，最近邻居几乎没有在其上打分
  
#### 基础解决方案
- 最好用皮尔森相似度
- 考虑共同打分物品的数目
- 对打分进行归一化处理
- 设置相似度阈值

#### 冷启动解决方法
1. 引导用户表达一些属性
2. 利用用户注册信息
3. 利用开放数据平台
4. 推荐排行榜单

#### 基于用户的方案为什么不流行
- 稀疏问题
- 数百万用户的相似度计算量太大
- 人对物品的喜欢程度会变

## 基于物品的协同过滤--节点为物品
1. 对物品提取特征向量
2. 计算的是物品关于用户打分的相似度，而不是物品内在属性的相似度

#### 基于物品的协同过滤优势
1. 计算性能高，因为一般用户数目远大于物品数目
2. 不易给用户打标签，容易给物品打标签
3. 物品之间的相似度一般不会变，用户之间的相似度可能会变

#### 冷启动解决方法
1. 文本分析
2. 主题模型
3. 打标签
4. 推荐旁行榜单

## 隐语义模型
1. 个性化推荐
2. 用户和物品具有隐含关系
